```python
from sklearn.cluster import KMeans
plt.figure(1,figsize=(10,7))
plt.title("The elbow method")

for n in range(1,10):

    wcss = []
    k = n*3; #Number of K in nth iteration
    for i in range(1,k):
        kmeans = KMeans(n_clusters=i,init="k-means++",n_init=10,max_iter=300,random_state=0)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)
    
    plt.tight_layout() 
    plt.subplot(3,3,n)
    
    plt.plot(range(1,k),wcss)
    plt.title("K: upto "+str(k))
    
    plt.xlabel("No clusters")
    plt.ylabel("Wcss")


```

**Visualize for different no of cluster**

```python
import random
n = 10 #Number of cluster 
row = int(n/2)
col = 2
plt.figure(1,figsize=(10,15))
for no_of_cluster in range(2,n):
    kmeans = KMeans(n_clusters=no_of_cluster,init="k-means++",n_init=10,max_iter=300,random_state=0)
    y_pred = kmeans.fit_predict(X)  
    plt.tight_layout() 
    for i in range(0,no_of_cluster):
        plt.subplot(row,col,no_of_cluster-1)
        plt.title("Numner of cluster: "+str(no_of_cluster))
        plt.scatter(X[y_pred==i,0],X[y_pred==i,1],s=100, color = [(random.randint(1,10)*.1,random.randint(1,10)*.1,random.randint(1,10)*.1)],label="Cluster 1")
        
```
